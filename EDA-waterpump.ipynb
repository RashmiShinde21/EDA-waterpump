# Data transformation

```python
t1.info()
```

The feature construction year has 55 unique values.For ease of analysis we group them into bins of 10.i,e 1960-1970(1),1970-1980(2)....son on here 1960s means 10 years of span that is 1960-1970 

```python
t1['construction_year'].value_counts().tail(30)
```

```python
def cons(t1):
    if t1['construction_year']<=1960:
         return '1960s'
    elif t1['construction_year']>1960 and t1['construction_year']<=1970:
         return '1970s'
    elif t1['construction_year']>1970 and t1['construction_year']<=1980:
         return '1980s'
    elif t1['construction_year']>1980 and t1['construction_year']<=1990:
         return '1990s'
    elif t1['construction_year']>1990 and t1['construction_year']<=2000:
         return '2000'
    elif t1['construction_year']>2000 and t1['construction_year']<=2010:
         return '2010'
    else:
         return '2020'
     
```

```python
t1['construction_year']=t1.apply(lambda x:cons(x),axis=1)
```
